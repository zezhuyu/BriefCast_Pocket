```python
from constant.preference import topic_embedding, sub_labels_embedding
preference_vector = np.zeros(1024)
for k, v in user["preference"]['subtopics'].items():
    sub_preference_vector = np.zeros(1024)
    if len(v) == 0:
        preference_vector += topic_embedding[k]
    else:
        for sub in v:
            sub_preference_vector += sub_labels_embedding[k][sub]
        preference_vector += sub_preference_vector / len(v)
```

```python
if isinstance(preference_vector, np.ndarray):
    prev_day_vector = preference_vector.tolist()
user_db.update({"prev_day_vector": prev_day_vector}, Query().id == user_id)
```
